```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup as bs
from time import sleep
from urllib import request
import requests
import re
import pandas as pd
import time
```

```python
billboard = pd.read_csv("TOP 100 Eng.csv")
url = "https://www.azlyrics.com/lyrics/{}/{}.html"
file = "lyrics_finalised_final_try_9.txt"

for index, row in billboard.iterrows():
    artist = row['artistNamethe']
    song = row['SongNameClean']
    artist = artist.replace(" ","")
    song.replace(' ', '')
    s = re.sub(r"\s","",song).lower() 
    

    try:
            
        with open(file,"a") as fi:
            f_url = url.format(artist,s)
            print(f_url)
            html_page = urllib.request.urlopen(f_url).read()
            soup = bs(html_page,'html.parser')
            html_pointer = soup.find('div', attrs={'class':'ringtone'})
            song_name = html_pointer.find_next('b').contents[0].strip()
            lyrics = html_pointer.find_next('div').text.strip()
            
            fi.write("==="+song_name+"===")
            fi.write("\n\n")
            fi.write(lyrics)
            
            fi.write("\n\n")
            time.sleep(3)
            print("Lyrics successfully written to file for : " + song_name)
            
    except:
        print(song_name+ "not found")
```
