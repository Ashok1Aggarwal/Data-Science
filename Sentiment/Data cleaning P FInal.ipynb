## Libraries

```python
import re
import pandas as pd
import numpy as np
import csv
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

## Cleaning unwanted characters

```python

file = open("Ashok rough.txt",encoding="utf8")
text = file.read()
text = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", " ", text.lower())
```

## Removing stop words like is, am, are

```python
stop = stopwords.words('english')
text = [word for word in text.split() if word not in (stop)]
```

## Counting word frequencies

```python
file = open("G:\SPOTIFY EDA\MyData\ALL CLEAN TEXT 1000.txt", 'r')
text = file.read()

def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

dict_data = word_count(text)
```

```python
csv_columns = ['Word', 'Frequency']
dict(dict_data)
```

## Converting Dictionary to CSV

```python
with open('WORD FRE.csv', 'w') as f:  # You will need 'wb' mode in Python 2.x
    w = csv.DictWriter(f, dict_data.keys())
    w.writeheader()
    w.writerow(dict_data)
```
